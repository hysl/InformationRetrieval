```python
import nltk
nltk.download('punkt')
nltk.download('wordnet')
```

```python
from urllib import request
url = "https://en.wikipedia.org/wiki/COVID-19_pandemic_in_New_York_(state)"
response = request.urlopen(url)
raw = response.read().decode('utf8')
```

```python
type(raw)
```

```python
len(raw)
```

```python
raw[:1000]
```

```python
from nltk import word_tokenize
tokens = word_tokenize(raw)
```

```python
type(tokens)
```

```python
len(tokens)
```

```python
tokens[:100]
```

```python
words = [w.lower() for w in tokens]
```

```python
words
```

```python
type(words)
```

```python
vocab = sorted(set(words))
```

```python
vocab
```

```python
type(vocab)
```

```python
porter = nltk.PorterStemmer()
lancaster = nltk.LancasterStemmer()
porter_tokens = [porter.stem(t) for t in tokens]
lancaster_tokens = [lancaster.stem(t) for t in tokens]
```

```python
porter_tokens
```

```python
lancaster_tokens
```

```python
wnl = nltk.WordNetLemmatizer()
wnl_tokens = [wnl.lemmatize(t) for t in tokens]
wnl_tokens
```

```python
output_file = open('vocab.txt', 'w')
words = set([w.lower() for w in wnl_tokens])
for word in sorted(words):
    print(word, file=output_file)
```
